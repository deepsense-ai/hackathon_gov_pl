```python
import codecs
import nltk
import os
from gensim import corpora, models
import pyLDAvis.gensim
import numpy as np
```

```python
DATA_PATH = "/data/hackathon/ustawy_pdf_demo/"
```

```python
all_files = []
for year in np.arange(1989, 2017):
    path = os.path.join(DATA_PATH, str(year))
    files = os.listdir(path)
    files = [os.path.join(str(year), f) for f in files if '.txt' in f]
    print(year, len(files))
    all_files.extend(files)
```

```python
def parse_file(filename, dir = DATA_PATH):
    fpath = os.path.join(dir, filename)
    with codecs.open(fpath, encoding="UTF-8") as f:
        content = f.read()
    return nltk.word_tokenize(content)
```

```python
#for f in files:
#    print(parse_file(f)[:100])
```

```python
texts = [parse_file(f) for f in all_files]
dictionary = corpora.Dictionary(texts)
dictionary.filter_extremes(no_above=0.3, no_below=20, keep_n=None) # TODO - no_below ustawiÄ‡ np na 10.
corpus = [dictionary.doc2bow(t) for t in texts]
```

```python
#with open('backup', 'wb') as f:
#    pickle.dump({'texts': texts, 'dictionary': dictionary, 'corpus': corpus}, f)
```

```python
%time lda_model = models.LdaMulticore(corpus, num_topics=20, id2word=dictionary, passes=50, workers=20)
```

```python
lda_vis_data = pyLDAvis.gensim.prepare(lda_model, corpus, dictionary)
pyLDAvis.display(lda_vis_data)
```
