```python
import codecs
import nltk
import os
from gensim import corpora, models
import pyLDAvis.gensimsim
```

```python
DATA_PATH = "/data/hackathon/tiny"
```

```python
files = os.listdir(DATA_PATH)
```

```python
def parse_file(filename, dir = DATA_PATH):
    fpath = os.path.join(dir, filename)
    with codecs.open(fpath, encoding="UTF-8") as f:
        content = f.read()
    return nltk.word_tokenize(content)
```

```python
for f in files:
    print(parse_file(f)[:100])
```

```python
texts = [parse_file(f) for f in files]

dictionary = corpora.Dictionary(texts)

dictionary.filter_extremes(no_above=0.3, no_below=1, keep_n=None) # TODO - no_below ustawić np na 10.

corpus = [dictionary.doc2bow(t) for t in texts]
```

```python
%time lda_model = models.LdaMulticore(corpus, num_topics=10, id2word=dictionary, passes=50, workers=10)
```

```python
lda_vis_data = pyLDAvis.gensim.prepare(lda_model, corpus, dictionary)
pyLDAvis.display(lda_vis_data)
```
